A drawback of the bottom-up approach in problem-solving or decision-making is that it may lack a comprehensive or holistic view of the problem or situation. b) Complete the code so it returns the number of rows that have at least one empty column value

c) Complete the code below to remove all columns with more than 200 NaN values

```python
# Remove all columns with more than 200 NaN values
df = df.dropna(axis=1, thresh=200)

# Print the remaining columns
print("Remaining columns:")
df.columns
```

d) Complete the code below to replaces `male` with 0 and `female` with 1

```python
# Replace 'male' with 0 and 'female' with 1 in the 'Sex' column
df['Sex'] = df['Sex'].replace({'male': 0, 'female': 1})

# Print the first few rows of the DataFrame to verify the changes
df.head()
```

e) Complete the code below to add four columns `First Name`, `Middle Name`, `Last Name`, and `Title` corresponding to the value in the `name` column.

For example: `Braund, Mr. Owen Harris` would be: 

|First Name | Middle Name | Last Name | Title |
|-----------|-------------|-----------|-------|
| Owen      |  Harris     |  Braund   | Mr    |

Anything not clearly one of the above 4 categories can be ignored.

```python
## Extracting the name components
name_parts = df['Name'].str.split(', |\. ', expand=True)

# Adding the extracted components as new columns
df['Title'] = name_parts[1]
df['First Name'] = name_parts[2]
first_name_split = df['First Name'].str.split(" ", expand=True)
df['First Name'] = first_name_split[0]
df['Middle Name'] = first_name_split[1]
df['Last Name'] = name_parts[0]


# Print the first few rows of the DataFrame to verify the changes
df.head()
```

f) Complete the code below to replace all missing ages with the average age

```python
# Calculate the average age
average_age = df['Age'].mean()

# Replace missing ages with the average age
df['Age'].fillna(average_age, inplace=True)

# Print the first few rows of the DataFrame to verify the changes
df.head()
```

g) Plot a bar chart of the average age of those that survived and did not survive. Briefly comment on what you observe.

```python
import matplotlib as plt
# Group the data by the 'Survived' column and calculate the average age for each group
average_age_by_survived = df.groupby('Survived')['Age'].mean()

# Plot a bar chart
average_age_by_survived.plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Age of Passengers by Survival')
plt.xlabel('Survived')
plt.ylabel('Average Age')
plt.xticks([0, 1], ['Not Survived', 'Survived'], rotation=0)
plt.show()
```
